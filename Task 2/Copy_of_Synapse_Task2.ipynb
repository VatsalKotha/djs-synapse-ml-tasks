## Task 2.2 - Operations

Write a code to create a list of all the movies that have the word 'Avengers' in it.

```python
# Your output will contain 2 movies, if correct
new_data = filtered_data[filtered_data['type'] == 'Movie'][filtered_data['title'].str.contains('Avengers')]['title'].tolist()
new_data
```

Display the number of Movies/TV shows per country.

```python
count_per_country = filtered_data.groupby(['type', 'country']).size().reset_index(name='count')

# Separate Movies and TV Shows
movies_per_country = count_per_country[count_per_country['type'] == 'Movie']
tv_shows_per_country = count_per_country[count_per_country['type'] == 'TV Show']

print("Movies per country:")
for index, row in movies_per_country.iterrows():
    print(f"Movie in {row['country']}: {row['count']}")

print("\nTV Shows per country:")
for index, row in tv_shows_per_country.iterrows():
    print(f"TV Show in {row['country']}: {row['count']}")
```

Display a new dataframe, that contains only those movies that are present in the Top 5 countries.

(Don't forget to reset the indices again)

```python
# The length of new dataframe will be 3102, if correct
movies_and_shows = filtered_data[filtered_data['type'].isin(['Movie', 'TV Show'])]

movie_show_counts_per_country = movies_and_shows['country'].value_counts()

top_countries = movie_show_counts_per_country.head(5).index

top_movies_shows = movies_and_shows[movies_and_shows['country'].isin(top_countries)]

top_movies_shows.reset_index(drop=True, inplace=True)

print(top_movies_shows.shape)
```

## Task 2.3 - Chart

*Use* the new dataframe to create a Pie Chart on the basis of number of Movies/TV Shows per country.

```python
counts = top_movies_shows['country'].value_counts()

plt.figure(figsize=(10, 6))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Movies/TV Shows per Country')
plt.axis('equal')
plt.show()
```
